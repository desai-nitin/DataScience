### Cleaning and Preparing the data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Problem Statement
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
#importing the dataset
df_train = pd.read_csv("black_friday/train.csv")
print(df_train.shape)
df_train.head()
```

```python
#load test data
df_test = pd.read_csv("black_friday/test.csv")
print(df_test.shape)
df_test.head()
```

```python
#Append both train and test data
df = df_train.append(df_test)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# UserId column is not useful so dropping it
df.drop(["User_ID"],axis=1,inplace=True)
df.head()
```

```python
## Handling Categorical Feature Gender
df["Gender"]=df["Gender"].map({'F':0,'M':1})
#pd.get_dummies(df["Gender"])
```

```python
df.head()
```

```python
## Handling Categorical Feature Age
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
print(df['Age'].unique())
```

```python
# Handling categorical feature city_category
#drop_first=True useful bcoz it reduces the number of columns, 
#when all the other columns are zero that means the first columns is 1.
df_cities=pd.get_dummies(df['City_Category'],drop_first=True)
df_cities
```

```python
df = pd.concat([df,df_cities],axis=1,verify_integrity=True)#ignore_index=True
df.head()
```

```python
# Dropping City_Category
df.drop(["City_Category"],axis=1,inplace=True)
df.head()
```

```python
# Handling Missing Values
df.isnull().sum()
```

```python
df["Product_Category_2"].unique()
```

```python
# Focus of replacing missing values
df["Product_Category_2"].value_counts()
```

As *Product_Category_2* is **Discrete variable** we will replace  Null Values with **Mode**

Not using **Mean** because Mean can give new value as category which we don't want


```python
df['Product_Category_2'].mode()[0]
```

```python
df["Product_Category_2"] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df["Product_Category_3"].unique()
```

```python
df["Product_Category_3"].value_counts()
```

```python
df['Product_Category_3'].mode()[0]
```

```python
df["Product_Category_3"] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df["Stay_In_Current_City_Years"].unique()
```

```python
df["Stay_In_Current_City_Years"]=df["Stay_In_Current_City_Years"].str.replace("+","")
```

```python
df.head()
```

```python
## Convert Object into integers
df["Stay_In_Current_City_Years"]=df["Stay_In_Current_City_Years"].astype(int)
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
df.info()
```
